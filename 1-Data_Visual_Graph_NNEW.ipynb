# Plot 2x3 Graphs: 
### 1. Scatterplot
### 2. Barplot
### 3. Boxplot
### 4. Time series plot
### 5. Time-based plot
### 6. Plot a function

```python
# Run this code
# Examples

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Data
np.random.seed(42)  # Đảm bảo kết quả ngẫu nhiên tái hiện được
x = np.array([-3, -2, -1, 0, 1, 2, 3, 4])
y = np.array([-3, -2, -1, 0, 1, 2, 3, 4])
my_ts = np.random.normal(size=500)  # Time series data
my_dates = pd.date_range(start="2005-01-01", periods=50, freq="M")
my_factor = pd.Series([6, 4, 8, 4, 6, 8, 4, 6] * 4, name="Cylinders")  # Tương tự mtcars$cyl

# Tạo lưới đồ thị 2x3
fig, axs = plt.subplots(2, 3, figsize=(15, 10))

# 1. Scatterplot
axs[0, 0].scatter(x, y, color='red')
axs[0, 0].set_title("Scatterplot")

# 2. Barplot
sns.countplot(x=my_factor, ax=axs[0, 1], palette="Reds")
axs[0, 1].set_title("Barplot")

# 3. Boxplot
sns.boxplot(x=my_factor, y=np.random.normal(size=len(my_factor)), ax=axs[0, 2])
axs[0, 2].set_title("Boxplot")

# 4. Time series plot
axs[1, 0].plot(np.arange(1, 501), my_ts, color='blue')
axs[1, 0].set_title("Time series")
axs[1, 0].set_xlabel("Time")
axs[1, 0].set_ylabel("Value")

# 5. Time-based plot
axs[1, 1].plot(my_dates, np.random.normal(size=50), marker='o')
axs[1, 1].set_title("Time based plot")
axs[1, 1].set_xlabel("Date")
axs[1, 1].set_ylabel("Value")

# 6. Plot a function
x_fun = np.linspace(0, 10, 100)
y_fun = x_fun**2
axs[1, 2].plot(x_fun, y_fun, color='green')
axs[1, 2].set_title("Plot a function")
axs[1, 2].set_xlabel("x")
axs[1, 2].set_ylabel("f(x) = x^2")

# Hiển thị đồ thị
plt.tight_layout()
plt.show()
```

# III.Density plot

### A Density Plot is a data visualization tool used to estimate the probability density function (PDF) of a continuous variable. 
### It shows the distribution of data and is smoother than a histogram, making it useful for identifying patterns, trends, and comparing distributions between datasets.
###

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import gaussian_kde

# Đọc dữ liệu mtcars từ file CSV
mtcars = pd.read_csv('mtcars.csv')

# Lấy cột 'mpg' (mile per gallon)
mile_per_gallon = mtcars['mpg']

# Vẽ histogram với xác suất (density=True tương đương prob = TRUE trong R)
plt.hist(mile_per_gallon, bins=6, density=True, color='blue', alpha=1.0, edgecolor='black', label='Histogram')

# Tính density và vẽ đường density
density = gaussian_kde(mile_per_gallon)
x_vals = np.linspace(mile_per_gallon.min(), mile_per_gallon.max(), 1000)
plt.plot(x_vals, density(x_vals), color='red', linewidth=6, label='Density')

# Thêm tiêu đề và nhãn trục
plt.title("Density histogram", fontsize=18)
plt.xlabel("milePerGallon", fontsize=14)
plt.ylabel("Density", fontsize=14)

# Hiển thị biểu đồ
plt.show()
```

# Vẽ density plot với hai băng thông (bandwidth) khác nhau sử dụng Matplotlib và Seaborn.

## Mục đích: 
### Density plot với các băng thông khác nhau giúp dễ dàng khám phá các xu hướng tiềm ẩn và bất thường trong dữ liệu, hỗ trợ trong phân tích thống kê và xây dựng mô hình.
###

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import gaussian_kde

# Đọc dữ liệu mtcars từ file
mtcars = pd.read_csv('mtcars.csv')

# Tạo dữ liệu ngẫu nhiên giống R
np.random.seed(0)
x = np.random.normal(size=500)

# Tạo layout 1 hàng, 2 cột
fig, axes = plt.subplots(1, 2, figsize=(12, 6))

# Đồ thị 1: Bandwidth lớn
density_big_bw = gaussian_kde(x, bw_method=20 / np.std(x))  # bw = 20
x_vals = np.linspace(min(x) - 50, max(x) + 50, 1000)
axes[0].plot(x_vals, density_big_bw(x_vals), color='red', linewidth=2)
axes[0].set_title("Too big bandwidth", fontsize=14, fontweight='bold', pad=20)
axes[0].set_xlabel("x", fontsize=12)
axes[0].set_ylabel("Density", fontsize=12)
axes[0].set_xlim(-60, 60)
axes[0].set_ylim(0, 0.02)
axes[0].text(0.5, -0.15, "N = 500    Bandwidth = 20", ha='center', fontsize=10, transform=axes[0].transAxes)

# Đồ thị 2: Bandwidth nhỏ
density_small_bw = gaussian_kde(x, bw_method=0.05 / np.std(x))  # bw = 0.05
axes[1].plot(x_vals, density_small_bw(x_vals), color='red', linewidth=2)
axes[1].set_title("Too small bandwidth", fontsize=14, fontweight='bold', pad=20)
axes[1].set_xlabel("x", fontsize=12)
axes[1].set_ylabel("Density", fontsize=12)
axes[1].set_xlim(-3, 3)
axes[1].set_ylim(0, 0.5)
axes[1].text(0.5, -0.15, "N = 500    Bandwidth = 0.05", ha='center', fontsize=10, transform=axes[1].transAxes)

# Hiển thị đồ thị
plt.tight_layout()
plt.show()
```

# Vẽ nhiều đường mật độ (density curves) sử dụng Matplotlib và Scipy.
## Mục đích:
### Việc vẽ nhiều đường mật độ là một cách mạnh mẽ để phân tích, so sánh và diễn giải dữ liệu.
###

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import gaussian_kde

# Tạo dữ liệu x
np.random.seed(42)
x = np.random.normal(size=500)

# Tạo dữ liệu y
np.random.seed(2)
y = np.random.normal(size=500) + 1

# Tính mật độ cho x
density_x = gaussian_kde(x)

# Tính mật độ cho y
density_y = gaussian_kde(y)

# Tạo lưới giá trị x để vẽ mật độ
x_grid = np.linspace(min(min(x), min(y)), max(max(x), max(y)), 1000)

# Vẽ đồ thị
plt.figure(figsize=(8, 6))
plt.plot(x_grid, density_x(x_grid), color='red', linewidth=2, label='Density of x')
plt.plot(x_grid, density_y(x_grid), color='blue', linewidth=2, label='Density of y')
plt.title("Multiple curves", fontsize=16)
plt.xlabel("", fontsize=12)
plt.legend()
plt.show()
```

# Vẽ nhiều đường mật độ (density curves) với giới hạn trục x và y được điều chỉnh chính xác

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import gaussian_kde

# Tạo dữ liệu
np.random.seed(42)
x = np.random.normal(size=500)

np.random.seed(2)
y = np.random.normal(size=500) + 1

# Tính mật độ
density_x = gaussian_kde(x)
density_y = gaussian_kde(y)

# Lưới giá trị cho trục x
x_grid_x = np.linspace(min(x), max(x), 1000)
x_grid_y = np.linspace(min(y), max(y), 1000)

# Xác định giới hạn trục x và y
x_min = min(x.min(), y.min())
x_max = max(x.max(), y.max())
y_min = min(density_x(x_grid_x).min(), density_y(x_grid_y).min())
y_max = max(density_x(x_grid_x).max(), density_y(x_grid_y).max())

# Vẽ đồ thị
plt.figure(figsize=(8, 6))
plt.plot(x_grid_x, density_x(x_grid_x), color='red', linewidth=2, label='Density of x')
plt.plot(x_grid_y, density_y(x_grid_y), color='blue', linewidth=2, label='Density of y')
plt.title("Multiple curves with correct axis limits", fontsize=16)
plt.xlabel("", fontsize=12)
plt.xlim(x_min, x_max)  # Giới hạn trục x
plt.ylim(y_min, y_max)  # Giới hạn trục y
plt.legend()
plt.show()
```

# Shade area under curve with transparency
## Mục đích:
### Giúp tăng cường độ rõ ràng, dễ đọc và khả năng diễn giải của biểu đồ, làm cho chúng hiệu quả hơn trong việc truyền tải thông tin và phân tích.
###

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import gaussian_kde

# Tạo dữ liệu x và y
np.random.seed(0)
x = np.random.normal(size=500)
y = np.random.normal(size=500) + 1

# Tính mật độ (density) cho x và y
density_x = gaussian_kde(x)
density_y = gaussian_kde(y)
x_vals = np.linspace(-4, 6, 1000)
y_vals = np.linspace(-4, 6, 1000)

# Vẽ biểu đồ
plt.figure(figsize=(10, 5))

# Biểu đồ đầu tiên: density của x và y với vùng tô màu
plt.subplot(1, 2, 1)
plt.plot(x_vals, density_x(x_vals), color='red', linewidth=2, label='Density of x')
plt.fill_between(x_vals, density_x(x_vals), color='red', alpha=0.5)
plt.plot(y_vals, density_y(y_vals), color='blue', linewidth=2, label='Density of y')
plt.fill_between(y_vals, density_y(y_vals), color='blue', alpha=0.5)
plt.xlim(-4, 6)
plt.ylim(0, 0.5)
plt.xlabel("")
plt.ylabel("Density")
plt.title("", fontsize=14)
plt.legend(loc='upper right')
plt.grid(True)

# Hiển thị đồ thị
plt.tight_layout()
plt.show()
```

# Making a Frequency table of Column

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Đọc dữ liệu mtcars từ file
mtcars = pd.read_csv('mtcars.csv')

# Tạo bảng tần suất tuyệt đối cho số xi-lanh (cột 'cyl')
my_table = mtcars['cyl'].value_counts().sort_index()

# Bố cục 1 hàng, 2 cột
plt.figure(figsize=(12, 6))

# Biểu đồ tần suất tuyệt đối
plt.subplot(1, 2, 1)
plt.bar(my_table.index.astype(str), my_table.values, color=['red', 'green', 'blue'])
plt.title("Absolute frequency", fontsize=16, fontweight='bold', pad=20)
plt.xlabel("Number of cylinders", fontsize=14)
plt.ylabel("Frequency", fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

# Biểu đồ tần suất tương đối
plt.subplot(1, 2, 2)
relative_freq = (my_table / my_table.sum()) * 100
plt.bar(relative_freq.index.astype(str), relative_freq.values, color=['red', 'green', 'blue'])
plt.title("Relative frequency (%)", fontsize=16, fontweight='bold', pad=20)
plt.xlabel("Number of cylinders", fontsize=14)
plt.ylabel("Percentage", fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

# Hiển thị đồ thị
plt.tight_layout()
plt.show()
```

# Text can be written on the bar plot as below

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Đọc dữ liệu mtcars từ file
mtcars = pd.read_csv('mtcars.csv')

# Tạo bảng tần suất của số xi-lanh (cyl)
my_table = mtcars['cyl'].value_counts().sort_index()

# Vẽ biểu đồ cột
colors = ['red', 'green', 'blue']  # Màu sắc theo thứ tự giống R
bar_positions = np.arange(len(my_table))  # Vị trí các cột
plt.bar(bar_positions, my_table.values, color=colors, edgecolor='black')

# Thêm nhãn vào các cột
for i, value in enumerate(my_table.values):
    plt.text(bar_positions[i], value + 0.2, str(value), ha='center', fontsize=12)

# Cài đặt trục y
plt.ylim(0, 15)

# Thêm tiêu đề và nhãn trục
plt.title("Bar Plot with Labels", fontsize=16)
plt.xlabel("Number of Cylinders", fontsize=14)
plt.ylabel("Frequency", fontsize=14)

# Đặt nhãn trục x
plt.xticks(bar_positions, my_table.index.astype(str), fontsize=12)

# Hiển thị biểu đồ
plt.show()
```

# Barplot with Legend

```python
import pandas as pd
import matplotlib.pyplot as plt

# Đọc dữ liệu mtcars từ file
mtcars = pd.read_csv('mtcars.csv')

# Tạo bảng tần suất của số xi-lanh (cyl)
my_table = mtcars['cyl'].value_counts().sort_index()

# Màu sắc cho các cột (tương tự rainbow(3) trong R)
colors = ['red', 'green', 'blue']

# Vẽ biểu đồ cột
bars = plt.bar(my_table.index.astype(str), my_table.values, color=colors, edgecolor='black')

# Thêm chú thích (legend) phía trên
plt.legend(bars, [f"{cyl} cylinders" for cyl in my_table.index], loc='upper center', bbox_to_anchor=(0.5, 1.15), ncol=3)

# Thêm nhãn trục và tiêu đề
plt.xlabel("Number of cylinders", fontsize=12)
plt.ylabel("Frequency", fontsize=12)
plt.title("Barplot with Legend on Top", fontsize=14)

# Hiển thị biểu đồ
plt.show()
```

# Grouping Bar Plot

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Tải dataset mtcars.csv từ file bên ngoài
mtcars = pd.read_csv('mtcars.csv')

# Chuyển đổi cột 'am' thành dạng phân loại
mtcars['am'] = mtcars['am'].map({0: "Automatic", 1: "Manual"})

# Tạo bảng tần suất giữa 'cyl' và 'am'
other_table = pd.crosstab(mtcars['cyl'], mtcars['am'])

# Màu sắc tùy chỉnh
colors = ["darkgrey", "darkblue", "red"]

# Vẽ barplot nhóm
bar_width = 0.25  # Chiều rộng của cột
x = np.arange(len(other_table.columns))  # Vị trí các nhóm (Automatic, Manual)

fig, ax = plt.subplots()

# Lặp qua từng số xi-lanh và vẽ các nhóm
for i, (cyl, values) in enumerate(other_table.iterrows()):
    ax.bar(x + i * bar_width, values, bar_width, 
           label=f"{cyl}", color=colors[i], edgecolor="black")

# Thêm tiêu đề và nhãn trục
ax.set_title("Grouped barchart", fontsize=14)
ax.set_xlabel("Transmission type", fontsize=12)
ax.set_ylabel("Frequency", fontsize=12)
ax.set_xticks(x + bar_width)  # Điều chỉnh nhãn trục x để nằm giữa
ax.set_xticklabels(other_table.columns, fontsize=10)

# Thêm legend
ax.legend(fontsize=10, title_fontsize=11, loc="upper right")

# Hiển thị biểu đồ
plt.tight_layout()
plt.show()
```

# V.Pie Chart.
### Pie or circular chart are used to represents discrete or categorical data. 
### The proportions or percentages in slices shows the quantities of various categories. 
### The pie() function is used to implement it

```python
import matplotlib.pyplot as plt

# Tạo một danh sách chứa số lượng của mỗi phần trong Pie Chart
count = [10, 20, 30, 40]

# Màu sắc tương tự hình đính kèm
colors = ['white', 'lightblue', 'lightcoral', 'lightcyan']

# Vẽ Pie Chart đơn giản với viền
plt.pie(count, 
        labels=[1, 2, 3, 4], 
        colors=colors, 
        wedgeprops={'edgecolor': 'black', 'linewidth': 1})

# Thêm tiêu đề
plt.title("Simple Pie Chart", fontsize=14)

# Hiển thị biểu đồ
plt.show()

```

```python
import matplotlib.pyplot as plt

# Dữ liệu cho Pie Chart
count = [10, 20, 30, 40]

# Tạo layout 1 hàng, 3 cột
fig, axes = plt.subplots(1, 3, figsize=(12, 4))  # figsize để điều chỉnh kích thước tổng thể

# Pie Chart 1: Nhãn và kích thước chữ lớn hơn
axes[0].pie(count, labels=count, colors=['white', 'lightblue', 'lightcoral', 'lightcyan'],
            wedgeprops={'edgecolor': 'black', 'linewidth': 1}, textprops={'fontsize': 10})
axes[0].set_title("Pie Chart 1", fontsize=14)

# Pie Chart 2: Màu cầu vồng, nhãn lớn hơn
axes[1].pie(count, labels=count, colors=plt.cm.rainbow(np.linspace(0, 1, 4)),
            wedgeprops={'edgecolor': 'black', 'linewidth': 1}, textprops={'fontsize': 12})
axes[1].set_title("Pie Chart 2", fontsize=14)

# Pie Chart 3: Màu cầu vồng (ít màu hơn), nhãn lớn hơn
axes[2].pie(count, labels=count, colors=plt.cm.rainbow(np.linspace(0, 1, 3)),
            wedgeprops={'edgecolor': 'black', 'linewidth': 1}, textprops={'fontsize': 10})
axes[2].set_title("Pie Chart 3", fontsize=14)

# Hiển thị biểu đồ
plt.tight_layout()
plt.show()

```

```python
import matplotlib.pyplot as plt

# Dữ liệu cho Pie Chart
count = [10, 20, 30, 40]
labels = [f"{c}%" for c in count]  # Tạo nhãn với dấu phần trăm

# Màu sắc giống hình đính kèm
colors = ['white', 'lightblue', 'lightcoral', 'lightcyan']

# Vẽ Pie Chart với nhãn phần trăm
plt.pie(count, 
        labels=labels, 
        colors=colors, 
        autopct=None,  # Không hiển thị thêm phần trăm tự động
        wedgeprops={'edgecolor': 'black', 'linewidth': 1})  # Thêm viền cho Pie Chart

# Thêm tiêu đề
plt.title("Pie Chart with Percentage Labels", fontsize=14)

# Hiển thị biểu đồ
plt.show()
```

```python
import matplotlib.pyplot as plt

# Dữ liệu cho Pie Chart
count_2 = [27, 15, 10, 25]

# Tạo nhãn tùy chỉnh với giá trị và phần trăm
total = sum(count_2)
pie_labels = [f"{value} = {value / total * 100:.2f}%" for value in count_2]

# Màu sắc tùy chỉnh
colors = ['lightblue', 'white', 'lightcyan', 'lightcoral']

# Vẽ Pie Chart với nhãn tùy chỉnh
plt.pie(count_2, 
        labels=pie_labels, 
        colors=colors, 
        wedgeprops={'edgecolor': 'black', 'linewidth': 1})  # Thêm viền đen cho các phần

# Thêm tiêu đề
plt.title("Pie Chart with Custom Labels", fontsize=14)

# Hiển thị biểu đồ
plt.show()
```
